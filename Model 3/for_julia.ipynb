```python
import os

# Get the list of files in the current directory
files = os.listdir()

# Print the list of files
print("List of files in the current directory:")
for file in files:
    print(f"'{file}',")
```

```python
csv_filenames = ['cities_data.csv',
'fishingground_location.csv',
'fishingground_restriction.csv',
'fishing_cost.xlsx',
'results.csv',
'fish_data.csv',
'for_julia.ipynb',
'processing_facility.csv',
'transportation_cost.xlsx',
'optimal_solution_for_julia.xlsx', 
]

for csv in csv_filenames:
    print(csv) 
    if csv[-3:] == 'csv':
        df = pd.read_csv(csv)
        if csv == 'fishingground_restriction.csv' or csv == 'optimal_x.csv': 
            df = df.round(2)
            latex_table = df.to_latex(index=False, longtable=True, float_format="%.2f")
            with open(f"Data//{csv}.tex", 'w') as f:
                f.write(latex_table)    
        else:    
            latex_table = df.to_latex(index=False, longtable=True)
            with open(f"Data//{csv}.tex", 'w') as f:
                f.write(latex_table)
    elif csv[-4:] == 'xlsx':
        xls = pd.ExcelFile(csv)
        count = 0
        for sheet_name in xls.sheet_names:
            count += 1
            df = pd.read_excel(xls, sheet_name)
            if csv == 'optimal_solution_for_julia.xlsx':
                df = df.round(2)
                display(df)
                latex_table = df.to_latex(index=False, longtable=True, float_format="%.2f")
                with open(f"Data//{csv}_{count}.tex", 'w') as f:
                    f.write(latex_table)
            else:
                latex_table = df.to_latex(index=False, longtable=True)
                with open(f"Data//{csv}_{count}.tex", 'w') as f:
                    f.write(latex_table)
            
```
